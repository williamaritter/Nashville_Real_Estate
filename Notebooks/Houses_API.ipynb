```python
import requests

url = "https://zillow-com1.p.rapidapi.com/propertyExtendedSearch"

querystring = {"location":"nashville, tn","home_type":"Houses"}

headers = {
	"X-RapidAPI-Host": "zillow-com1.p.rapidapi.com",
	"X-RapidAPI-Key": ""
}
```

```python
all_houses = pd.DataFrame()
for page in range (10, 20):
    querystring = {"location":"nashville, tn","home_type":"Houses", "status_type":"ForSale", "page":page}
    response = requests.get(url, headers=headers, params=querystring)
    print (response)
    for prop in range (0,40):
        house = pd.DataFrame(response.json()["props"][prop])
        all_houses = pd.concat([all_houses, house])
        
all_houses
```
