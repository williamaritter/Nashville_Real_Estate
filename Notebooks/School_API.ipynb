   # Read in Schools as "Each_House" DF

```python
each_house = pd.DataFrame()
zpid_lists = []
level = []
rating = []

for item in range (0, len(all_data)):
    test_id = all_data.loc[item, "zpid"]
    querystring = {"zpid":test_id} 
    print (querystring)
    response = requests.get(url, headers = headers, params = querystring)

    if "schools" in list(response.json().keys()):

        a_house = pd.DataFrame(response.json()["schools"])
        a_house["zpid"] = test_id
        each_house = pd.concat([each_house, a_house])

        for i in range(0, len(response.json()["schools"])):
            zpid_lists.append(test_id)
            level.append(response.json()["schools"][i]["level"])
            rating.append(response.json()["schools"][i]["rating"])
            time.sleep(1)
            
    
    

```
